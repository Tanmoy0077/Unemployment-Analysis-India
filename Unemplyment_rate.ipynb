# Data Analysis on Unemployment Rate of all the states in India

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Datasets/Unemployment in India.csv', parse_dates=['Date'], dayfirst=True)
```

```python
df.head()
```

```python
df.drop(['Frequency'], axis=1, inplace=True)
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
i = 1
plt.figure(figsize=(30,30))
sns.set_style('darkgrid')
for reg in df.Region.unique():
    plt.subplot(7, 4, i)
    temp = df[df['Region'] == reg]
    sns.lineplot(data=temp, x='Date', y='Estimated Unemployment Rate (%)', hue='Area', palette='winter')
    plt.ylabel(f'Estimated Unemployment Rate (%) in {reg}')
    i+=1
plt.savefig('Unemployment Rate in all Regions of India.png')
```

```python
i = 1
sns.set(font_scale=0.8)
plt.figure(figsize=(30,30))
sns.set_style('darkgrid')
for reg in df.Region.unique():
    plt.subplot(7, 4, i)
    temp = df[df['Region'] == reg]
    sns.lineplot(data=temp, x='Date', y='Estimated Employed', hue='Area', palette='rocket')
    plt.ylabel(f'Estimated Employed in {reg}')
    i+=1
```

```python
i = 1
sns.set(font_scale=0.8)
plt.figure(figsize=(30,30))
sns.set_style('darkgrid')
for reg in df.Region.unique():
    plt.subplot(7, 4, i)
    temp = df[df['Region'] == reg]
    sns.lineplot(data=temp, x='Date', y='Estimated Labour Participation Rate (%)', hue='Area', palette='summer')
    plt.ylabel(f'Estimated Labour Participation Rate (%) in {reg}')
    i+=1
```

```python
plt.figure(figsize=(20,15))
sns.set(font_scale=2)
sns.barplot(data = df[df.Date == '2020-06-30'], x='Estimated Employed', y='Region', hue='Area', palette='crest', orient='h')
plt.title('Estimated Employed on 2020-06-30 in all Regions')
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15,6))
sns.kdeplot(data=df,x='Estimated Unemployment Rate (%)', hue='Area', fill=True)
```

```python
plt.figure(figsize=(15,6))
sns.kdeplot(data=df,x='Estimated Employed', hue='Area', fill=True)
```

```python
plt.figure(figsize=(15,6))
sns.kdeplot(data=df,x='Estimated Labour Participation Rate (%)', hue='Area', fill=True)
```

```python
urban_df = df[df.Area == 'Urban'].pivot(index='Region', columns='Date', values='Estimated Unemployment Rate (%)')
```

```python
rural_df = df[df.Area == 'Rural'].pivot(index='Region', columns='Date', values='Estimated Unemployment Rate (%)')
```

```python
plt.figure(figsize=(15,10))
sns.set(font_scale=0.8)
xlabels = ['2019-05', '2019-06','2019-07','2019-08','2019-09','2019-10','2019-11','2019-12',
                  '2020-01','2020-02','2020-03','2020-04','2020-05','2020-06']
sns.heatmap(urban_df, cmap='Dark2', linewidths=0.8, annot=True, xticklabels=xlabels)
```

```python
plt.figure(figsize=(15,10))
sns.set(font_scale=0.8)
sns.heatmap(rural_df, cmap='Dark2', linewidths=0.8, annot=True, xticklabels=xlabels)
```
